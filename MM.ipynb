1. Berikut nomor satu

Disini kami menggunakan Algoritma untuk mencari fibonacci sequence(O(2^n)). Jadi kami menggunakan algoritma ini karena ini masih termasuk algoritma dasar yang memiliki fungsi pertumbuhan dimana jumlahnya akan terus bertambah sampai urutan terakhirnya. 

Fibonacci memiliki pola berikut: 0, 1, 1, 2, 3, 5, 8....dst.

Bisa dilihat dari polanya kalau angka pertama dan kedua terdiri atas 0 dan 1. Angka-angka selanjutnya diisi oleh jumlah kedua angka sebelumnya.

(xn) = (x-1) + (x-2)

```python
print("Berapa suku yang diperlukan?")
ketentuan = int(input())

# dua suku pertama
x1 = 0
x2 = 1
count = 0


# mengecek jika jumlah sukunya tidak negatif
if ketentuan <= 0:
   print("Tolong masukkan angka yang positif")

# Jika sukunya hanya berjumlah satu, return x1nya
elif ketentuan == 1:
   print("Fibonacci sequence sampai dengan", ketentuan ,":")
   print(x1)
   
# memprint fibonacci sequencenya
else:
   print("Fibonacci sequencenya:")
   while count < ketentuan:
       print(x1)
       xn = x1 + x2
       # mengupdate angkanya
       x1 = x2
       x2 = xn
       count += 1
        
```

```python
x = np.array([0,0.5,1,6])
y = np.array([0,-0.5,0,6])
X_Y_Spline = make_interp_spline(x, y)
X_ = np.linspace(x.min(), x.max(), 500)
Y_ = X_Y_Spline(X_)
fig = plt.figure()
ax = plt.axes()
plt.grid()
plt.plot(x, y)
plt.title("y = n log n")
plt.xlabel("sumbu x")
plt.ylabel("sumbu y")
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.show()
```
